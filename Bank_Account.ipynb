```python
import sys
```

```python
def check_balance(balance):
    print(f"Your current balance:${balance}")
    sys.stdout.flush()


def deposit_money(balance):
    # global balance
    while True:
        user_input = input("Enter deposit amount, or press 'q' to go back")
        if user_input == "q" or user_input == "Q":
            break
        try:
            amount = float(user_input)
            if amount > 0:
                balance += amount
                print(f"Done.Your current balance:${balance}")
                sys.stdout.flush()
                break
        except ValueError:
            print("Please enter a valid positive number")
            sys.stdout.flush()
            continue
    return balance


def withdraw_money(balance):
    # global balance
    while True:
        user_input = input("Enter withdrawal amount, or press 'q' to go back")
        if user_input == "q" or user_input == "Q":
            break
        try:
            amount = float(user_input)
            if amount > 0:
                if amount <= balance:
                    balance -= amount
                    print(f"Done.Your current balance:${balance}")
                    sys.stdout.flush()
                    break
                else:
                    print(f"Not enough balance.Your current balance:${balance}")
                    sys.stdout.flush()
        except ValueError:
            print("Please enter a valid positive number")
            sys.stdout.flush()
            continue
    return balance

balance = 0
while True:
    

    want = input(
        "Press 1 to check your balance, 2 to deposit money, 3 to withdraw money, and 4 to exit"
    )
    if want == "1":
        check_balance(balance)
    elif want == "2":
        balance = deposit_money(balance)
    elif want == "3":
        balance = withdraw_money(balance)
    elif want == "4":
        break
    else:
        print("Please select the correct option.")
        sys.stdout.flush()
        continue
```

```python
accountBalance = 0  # Initialize the account balance globally


def Deposit(balance, amount):
    return balance + amount


def Withdraw(balance, amount):
    if amount > balance:
        print("Insufficient funds.")
        sys.stdout.flush()
        return balance
    return balance - amount


def checkBalance(balance):
    return balance


while True:
    operation = input("Enter 1 = check balance, 2 = deposit, 3 = withdraw : ")

    if operation == "1":
        current_balance = checkBalance(accountBalance)
        print(f"Your balance is: {current_balance}")
        sys.stdout.flush()
    elif operation == "2":
        try:
            amount = float(input("Enter deposit amount: "))
            if amount > 0:
                accountBalance = Deposit(accountBalance, amount)
                print(f"Deposit successful. Current balance: {accountBalance}")
                sys.stdout.flush()
            else:
                print("Deposit amount must be positive.")
                sys.stdout.flush()
        except ValueError:
            print("Invalid input. Please enter a number.")
            sys.stdout.flush()
            continue
    elif operation == "3":
        try:
            amount = float(input("Enter withdrawal amount: "))
            if amount < accountBalance:
                accountBalance = Withdraw(
                    accountBalance, amount
                )  # Update global accountBalance
                print(f"Withdrawal successful. Current balance: {accountBalance}")
                sys.stdout.flush()
            else:
                print(
                    "Your withdrawal request exceeds your account balance. Please enter a new withdrawal amount."
                )
                sys.stdout.flush()
                while True:

                    amount = float(input("Enter the new amount: "))
                    try:
                        if amount > accountBalance:
                            print(
                                "Your withdrawal request exceeds your account balance. Please enter a new withdrawal amount."
                            )
                            sys.stdout.flush()
                        else:
                            accountBalance = Withdraw(accountBalance, amount)
                            print(
                                f"Withdrawal successful. Current balance: {accountBalance}"
                            )
                            sys.stdout.flush()
                            break
                    except ValueError:
                        print("Invalid input. Please enter a number.")
                        sys.stdout.flush()
                        continue

        except ValueError:
            print("Invalid input. Please enter a number.")
            sys.stdout.flush()
            continue

    otheroperation = str(
        input("Do you want to continue or not, Type yes to continue and no to exit")
    )
    if otheroperation == "yes":
        continue
    elif otheroperation == "no":
        break
```
